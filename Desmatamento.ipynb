```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('desmatamento_amazonia_estados_prodes_2012-2015.csv', sep = ';')
```

```python
df
```

```python
plt.axis('equal')
plt.pie(df['Estado'].value_counts())
plt.suptitle("Estados com maior índice de desmatamento 2012-2015")
plt.show()
print(df['Estado'].value_counts())
```

```python
plt.axis('equal')
plt.pie(df['Ano'].value_counts())
plt.suptitle("Quantidade de Desmatamento por Ano")
plt.show()
print(df['Ano'].value_counts())
```

```python
plt.axis('equal')
plt.pie(df['Categoria'].value_counts())
plt.suptitle("Categoria do Desmatamento")
plt.show()
print(df['Categoria'].value_counts())
```
